### ABSTRACT

In most existing methods of music mood classification, the moods of songs are divided according to psychologist Robert Thayer’s traditional model of mood. 

The model divides songs along the lines of energy and stress, from happy to sad and calm to energetic, respectively (Bhat et al 359). The eight categories created by Thayer’s model include the extremes of the two lines as well as each of the possible intersections of the lines (e.g. happy-energetic or sad-calm). Tempo: The speed of music, determined by the frequency of beats. Tempo strongly correlates with the perceived energy level of a piece of music.

Timbre: The tonal quality of a sound, influenced by harmonic components. Timbre analysis involves properties such as zero-crossing rate and spectral irregularity, which provide insights into the overall tone and saturation of a piece. Higher timbre values generally indicate higher energy in the music.

Pitch: The frequency of a sound, indicating whether it's high or low. Higher pitched music tends to be associated with positive moods like happiness and energy, while lower pitched music may evoke sadder or calmer emotions.

Intensity: The loudness or average volume of a song. Intensity is a strong indicator of a song's energy level. Higher intensity typically corresponds to higher energy, while lower intensity suggests a calmer piece.
